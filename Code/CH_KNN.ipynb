```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
ds=pd.read_csv('/Users/harshdixit/Downloads/Project_crop_health/Crop_recommendation.csv')
```

```python
ds.head()
```

```python
li=[]
for it in ds.label:
    if it not in li:
        li.append(it)
print("Number of labels/crop=",len(li))
print("Name of crops=",li)
```

```python
target=ds.label.astype('category').cat.codes
ds['Crop_Type_label']=target
```

```python
ds.head()
```

### adding column for soil health level (1-excellent to 5-worst)

```python
ds['Level'] = np.random.randint(1,6,ds.shape[0])
```

```python
ds.head()
```

```python
ds.describe()
```

```python
x=ds[['N','P','K','temperature','humidity','ph','rainfall','Crop_Type_label']]
y=ds.Level
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.4,random_state=0)
```

```python
#for different values of k
k=[3,4,5,6,7,8,9]
score=[]
```

```python
for it in k:
    knn=KNeighborsClassifier(n_neighbors=it)
    knn.fit(xtrain,ytrain)
    score.append(knn.score(xtest,ytest))
```
